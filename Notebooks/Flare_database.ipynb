```python
# import stuff
import os
import pandas as pd
import scipy.io as sio

# save data 
sdat = True
sdir = 'C:/Users/krmurph1/OneDrive/SatDrag_RF'

# Read in the satellite drag database
#fn = 'D:\\data\\VL_sdrag\\satdrag_database_zlib.hdf5'
fn = 'D:\\data\\SatDensities\\satdrag_database_grace_B.hdf5'
sd_dat = pd.read_hdf(fn)
```

```python
# Read in Flare Database
# and convert to DataFrame

fl_dat = sio.readsav("d:/data/flare list/GOES_FLARE_RECORDS_1975_2024.sav")
d = dict()
for key in fl_dat['goes'].dtype.names:                             
    v = fl_dat['goes'][key]            
    try:                                                                  
        v = v.byteswap().newbyteorder()
    except ValueError:
        pass
    d[key] = v

fl_df = pd.DataFrame(d)
```

```python
# loop through the flares and set
# flags in the satellite drag database
# for the different times

sd_dat['flare'] = -1
sd_dat['flare class'] = -1
sd_dat['flare magnitude'] = -1
sd_dat['flare peak energy'] = -1
sd_dat['flare index'] = -1


for index, row in fl_df.iterrows():

    st_time = pd.to_datetime(row['START_TIME'].decode())
    if st_time < sd_dat['DateTime'].min():
        continue

    try:        
        en_time = pd.to_datetime(row['END_TIME'].decode())
    except:
        print(f"Times: {row['START_TIME'].decode()} - {row['END_TIME'].decode()}")
        continue       

    

    fl_t = (sd_dat['DateTime'] >= st_time) & (sd_dat['DateTime'] <= en_time) 


    
    sd_dat.loc[fl_t,'flare'] = 1
    try:
        sd_dat.loc[fl_t,'flare class'] = row['CLASS'].decode()
    except:
        sd_dat.loc[fl_t,'flare class'] = row['CLASS']
    sd_dat.loc[fl_t,'flare magnitude'] = row['FLARE_MAGNITUDE']
    sd_dat.loc[fl_t,'flare peak energy'] = row['PEAK_ENERGY']
    sd_dat.loc[fl_t,'flare index'] = index
```
